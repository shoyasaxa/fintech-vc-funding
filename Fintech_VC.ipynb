```python
import pandas as pd
```

```python
funded_df = pd.read_csv("crunchbase_export.csv")
```

```python
funded_df.head(50)
```

```python
df_fintech = pd.DataFrame() 

exception_counter = 0
for index, row in funded_df.iterrows():
    try:
        if( ("fintech" in row["short_description"]) or ("fintech" in row["category_list"]) or ("fintech" in row["category_group_list"]) ): 
            df_fintech = df_fintech.append(row)
    except:
        exception_counter += 1 
        
print("there were {} exceptions".format(exception_counter))
df_fintech.head(100)
```

```python
print("there are {} fintech companies funded".format(len(df_fintech)))
```

```python
df_fintech.to_csv("Funded Fintech Companies")
```

```python
writer = pd.ExcelWriter("Funded-Fintech-Companies.xlsx")
df_fintech.to_excel(writer)
writer.save()
```
