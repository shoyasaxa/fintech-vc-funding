```python
import pandas as pd
```

```python
funded_df = pd.read_csv("crunchbase_export.csv")
```

```python
funded_df.head(5)
```

```python
df_fintech = pd.DataFrame() 

exception_counter = 0
for index, row in funded_df.iterrows():
    try:
        if( ("fintech" in row["short_description"]) or ("fintech" in row["category_list"]) or ("fintech" in row["category_group_list"]) ): 
            df_fintech = df_fintech.append(row)
    except:
        exception_counter += 1 
        
print("there were {} exceptions".format(exception_counter))
df_fintech.head(5)
```

```python
print("there are {} fintech companies funded".format(len(df_fintech)))
```

```python
country_counts = df_fintech["country_code"].value_counts()
display(country_counts.head(15))
```

```python
from matplotlib import pyplot as plt 

plt.pie(country_counts, labels=country_counts.index)
plt.show();
```

```python
city_counts = df_fintech["city"].value_counts()
city_counts.head(15)
```

```python
plt.pie(city_counts, labels=city_counts.index)
plt.show();
```

```python
df_fintech["last_funding_datetime"] = pd.to_datetime(df_fintech["last_funding_on"],format="%Y-%m-%d")
df_fintech.head(1)
```

```python
from datetime import datetime

date_2014_1_1 = datetime(2014,1,1)
date_2015_1_1 = datetime(2015,1,1)
date_2016_1_1 = datetime(2016,1,1)
date_2017_1_1 = datetime(2016,1,1)
date_2018_1_1 = datetime(2018,1,1)
date_ranges = [date_2014_1_1, date_2015_1_1, date_2016_1_1, date_2017_1_1, date_2018_1_1]
```

```python
for i in range(len(date_ranges)-1):
    if (i==0):
        df = df_fintech[df_fintech["last_funding_datetime"] < date_ranges[0]] 
        country_counts = df["country_code"].value_counts() 
        display(country_counts.head(15))
        plt.pie(country_counts,labels=country_counts.index)
        plt.plot(); 
        
        city_counts = df["city"].value_counts() 
        city_counts.head(15) 
        plt.pie(city_counts,labels=city_counts.index)
        plt.plot(); 
        
#     else:
#         df = df_fintech[(df_fintech["last_funding_datetime"] > date_ranges[i]) & (df_fintech["last_funding_datetime"] < date_ranges[i+1]) ]
#         country_counts = df["country_code"].value_counts() 
#         country_counts.head(15)
#         plt.pie(counts, labels=country_counts.index)
#         plt.plot(); 
        
#         city_counts = df["city"].value_counts() 
#         city_counts.head(15) 
#         plt.pie(city_counts,labels=city_counts.index)
#         plt.plot(); 
        
```
